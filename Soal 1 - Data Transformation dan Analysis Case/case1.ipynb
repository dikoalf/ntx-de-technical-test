# Task 1

Case : Identify top products based on the total transaction revenue per day.

```python
# menampilkan top produk setiap harinya dengan cara mengroup data bedasarkan date, product name dan revenuenya
topProducts = df.groupby(['date', 'v2ProductName'])['totalTransactionRevenue'].sum().reset_index().sort_values(by=['date', 'totalTransactionRevenue'], ascending=False)
topProducts
```

dari table kita dapat mengetahui produk apa yang menghasilkan revenue paling banyak setiap harinya. tapi karna banyak dari data yang tidak memiliki nilai, maka kita akan mengambil data yang hanya memiliki nilai. 

```python
# mengambil data dimana kolom totalTransactionRevenue memiliki nilai selain null
dfRevenue = df[['totalTransactionRevenue', 'v2ProductName', 'date']].dropna(subset=['totalTransactionRevenue'])

# menampilkan top produk setiap harinya dengan cara mengroup data bedasarkan date, product name dan revenuenya
topProducts = dfRevenue.groupby(['date', 'v2ProductName'])['totalTransactionRevenue'].sum().reset_index().sort_values(by=['date', 'totalTransactionRevenue'], ascending=False)
topProducts
```

setelah difilter table keliatan lebih jelas dan akurat, sehingga kita dapat dengan mudah mengetahui produk apa yang paling banyak menghasilkan revenue setiap harinya

# Task 2

Case : Detect any anomalies, such as a sharp decrease or increase in the number of transactions for a specific product.

```python
# Mengambil nama produk untuk digunakan saat visualisasi dan pengambilan spesifik produk
products = df['v2ProductName'].unique()

# looping untuk setiap produk
for name in products:
    data = df[df['v2ProductName'] == name].copy() 

    # menghitung Z-score
    data['zScore'] = (data['transactions'] - data['transactions'].mean()) / data['transactions'].std()

    # menandai anomali berdasarkan Z-score
    data['anomaly'] = data['zScore'].abs() > 3 

    # menjabarkan figursize
    fig, ax = plt.subplots(figsize=(20,5))

    # membuat plot visualnya
    sns.lineplot(data=data, x='date', y='transactions', ax=ax, label='Transactions')
    # menandai anomali
    sns.scatterplot(data=data[data['anomaly']], x='date', y='transactions', color='red', label='Anomalies', s=100)

    plt.title(f'Trend Products {name}')
    plt.legend()
    plt.tight_layout()
    plt.show()
```

Dengan line chart diatas kita dapat mengetahui trend setiap produk yang dijual, dari line chart pula kita dapat melihat volume transaksi produk tersebut

# Task 3

Case :
Identify the most profitable city or province based on the total transaction revenue.

```python
# mengelompokkan data berdasarkan city dan total transaction revenue
dfProfitableCity = df.groupby(['city'])['totalTransactionRevenue'].sum().reset_index()

# mengurutkan kota berdasarkan total transaction revenue secara menurun
dfProfitableCity = dfProfitableCity.sort_values(by='totalTransactionRevenue', ascending=False)

# menjabarkan figursize
fig, ax = plt.subplots(figsize=(20,20))

# membuat plot visualnya
sns.barplot(data=dfProfitableCity, x='totalTransactionRevenue', y='city', ax=ax, palette='viridis')

plt.xlabel('Total Transaction Revenue')
plt.ylabel('City')
plt.title('Total Transaction Revenue by City')
plt.tight_layout()
plt.show()
```

Dari sini, kita dapat melihat bahwa ‘not available in demo dataset’ adalah kota nomor 1. Kota ini bukanlah data asli, melainkan gabungan dari kota-kota yang belum tersedia saat demo. Langkah berikutnya adalah mengambil data selain ‘not available in demo dataset’ agar visualisasi lebih akurat.

```python
# mengambil data dimana kolom totalTransactionRevenue memiliki nilai selain null dan mengambil data dimana kolom city bukan berisi 'not available in demo dataset'
dfRevenue = df[['totalTransactionRevenue', 'v2ProductName', 'date', 'city']].dropna(subset=['totalTransactionRevenue'])
dfRevenue = dfRevenue[dfRevenue['city'] != 'not available in demo dataset']

# mengelompokkan data berdasarkan city dan total transaction revenue
dfProfitableCity = dfRevenue.groupby(['city'])['totalTransactionRevenue'].sum().reset_index()

# mengurutkan kota berdasarkan total transaction revenue secara menurun
dfProfitableCity = dfProfitableCity.sort_values(by='totalTransactionRevenue', ascending=False)

# menjabarkan figursize
fig, ax = plt.subplots(figsize=(20,20))

# membuat plot visualnya
sns.barplot(data=dfProfitableCity, x='totalTransactionRevenue', y='city', ax=ax, palette='viridis')

plt.xlabel('Total Transaction Revenue')
plt.ylabel('City')
plt.title('Total Transaction Revenue by City')
plt.tight_layout()
plt.show()
```

setelah mengambil data selain 'not avaiable in demo dataset' berhasil didapatkan kota paling menguntungkan secara akurat.
